### Reshape accident counts for trend analysis

The original dataset stores 2021–2024 accident counts in separate columns.
A long-format table is created so each State/UT and year has one observation,
making the data suitable for time-series charts and dashboard filters.

```python
accident_year_columns = [
    "accidents_2021",
    "accidents_2022",
    "accidents_2023",
    "accidents_2024"
]

trend_data = state_data.melt(
    id_vars=["serial_no", "state_ut", "region"],
    value_vars=accident_year_columns,
    var_name="year",
    value_name="accident_count"
)

display(trend_data.head(10))
print("Trend data shape:", trend_data.shape)
```

```python
trend_data["year"] = (
    trend_data["year"]
    .str.replace("accidents_", "", regex=False)
    .astype("Int64")
)

trend_data["accident_count"] = (
    pd.to_numeric(
        trend_data["accident_count"],
        errors="coerce"
    )
    .astype("Int64")
)

trend_data = trend_data.sort_values(
    ["state_ut", "year"]
).reset_index(drop=True)

display(trend_data.head(12))
print(trend_data.dtypes)
```

```python
print("Number of rows:", len(trend_data))
print("Number of State/UTs:", trend_data["state_ut"].nunique())
print("Years:", sorted(trend_data["year"].unique().tolist()))

print("\nMissing values:")
print(
    trend_data.isna().sum()[
        trend_data.isna().sum() > 0
    ]
)
```

### Data validation

State/UT accident counts are summed for each year and compared with the
national Total row from the source file. A difference of zero confirms
that no records were lost or duplicated during cleaning.

```python
year_columns = {
    2021: "accidents_2021",
    2022: "accidents_2022",
    2023: "accidents_2023",
    2024: "accidents_2024"
}

validation_rows = []

for year, column in year_columns.items():
    calculated_total = int(
        pd.to_numeric(
            state_data[column],
            errors="coerce"
        ).sum()
    )

    reported_total = int(
        pd.to_numeric(
            national_totals[column],
            errors="coerce"
        ).iloc[0]
    )

    validation_rows.append({
        "year": year,
        "calculated_state_total": calculated_total,
        "reported_national_total": reported_total,
        "difference": calculated_total - reported_total,
        "matches": calculated_total == reported_total
    })

validation_df = pd.DataFrame(validation_rows)

display(validation_df)
```

```python
assert validation_df["matches"].all(), (
    "Validation failed: State/UT totals do not match "
    "the reported national totals."
)

print("Validation passed: all yearly totals match.")
```

### Export cleaned datasets

Three processed files are exported:

1. `accidents_cleaned.csv` — State/UT-level dataset with rates, regions and derived metrics
2. `accidents_trend_long.csv` — long-format accident counts for trend charts
3. `national_totals.csv` — reported national totals kept separately for validation

```python
processed_dir = Path("../data/processed")
processed_dir.mkdir(parents=True, exist_ok=True)

cleaned_output_path = processed_dir / "accidents_cleaned.csv"
trend_output_path = processed_dir / "accidents_trend_long.csv"
totals_output_path = processed_dir / "national_totals.csv"

state_data = state_data.sort_values("serial_no").reset_index(drop=True)

state_data.to_csv(
    cleaned_output_path,
    index=False
)

trend_data.to_csv(
    trend_output_path,
    index=False
)

national_totals.to_csv(
    totals_output_path,
    index=False
)

print("Files exported successfully:")
print(cleaned_output_path.resolve())
print(trend_output_path.resolve())
print(totals_output_path.resolve())
```
